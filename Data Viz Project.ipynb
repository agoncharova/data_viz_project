## Area Chart 

This section of the project is dedicated to creating a stacked area chart, incorporating all four engine type categories – Gas, Petrol, Diesel, and Other.

```python
df_fuel_engine_types = pd.read_csv(r"stacked_area_chart_data.csv")
```

```python
df_fuel_engine_types
```

```python
colors = ['#7e2987', '#3498db', '#e74c3c', '#2ecc71']
labels = ["Diesel", "Petrol", "Gas", "Other"]
sns.set_style("white")
plt.figure(figsize = (12, 6))
plt.stackplot(df_fuel_engine_types["Year"],
              df_fuel_engine_types["Diesel"],
              df_fuel_engine_types["Petrol"],
              df_fuel_engine_types["Gas"],
              df_fuel_engine_types["Other"],
              colors = colors,
              edgecolor = 'none')
plt.xticks(df_fuel_engine_types["Year"], rotation = 45)
plt.legend(labels = labels, loc = "upper left")
plt.ylabel("Number of Cars", fontsize = 13)
plt.title("Popularity of Engine Fuel Types (1982 - 2016)", fontsize = 14, weight = "bold")
sns.despine()
plt.show()
```

In the presented chart, we can observe a notable trend in the popularity of different fuel types over the years. Since 1982, there has been a significant increase in the number of cars, with Diesel emerging as the most popular fuel type. Its popularity reached its peak around 2007-2008. In this period, we can also notice a hight rise in the number of cars using petrol as a fuel, almost on par with diesel cars. However, following the financial crisis in 2009, there was a noticeable decline in the number of cars as well as in fuel type usage. The second rise comes in 2012 with remaining dominance of diesel.

## Area Chart of a Company's Revenue

The goal of this part is to create a stacked area chart of a printer company and its revenue for the last 12 months. The data set contains monthly values in the span of a year and features three revenue sources:

-International Printer Sales

-Domestic Printer Sales

-Services (fixed fee)

```python
df_revenue = pd.read_csv(r"C:\Users\Настя\Downloads\area_chart_project_data.csv")
```

```python
df_revenue
```

```python
df_revenue['Date'] = pd.to_datetime(df_revenue['Date'])
```

```python
colors = ['#27ae60', '#218c74', '#1abc9c']
plt.figure(figsize = (12,6))
labels = ["International", "Services", "Domestic"]
sns.set_style("white")
plt.stackplot(df_revenue["Date"],
              df_revenue["International"],
              df_revenue["Services"],
              df_revenue["Domestic"],
              colors = colors,
              edgecolor = 'none')
plt.legend(labels = labels, loc = "upper left")
plt.xticks(df_revenue["Date"], rotation = 45)
plt.ylabel("Revenue in $", fontsize = 13)
plt.title("Revenue Report 2020", fontsize = 14, weight = "bold")
sns.despine()
plt.show()
```

The chart illustrates the revenue of the printer company for the year 2020, highlighting international revenue as the most substantial, followed by services, and the smallest contribution from domestic sources.

Three distinct peaks are evident at the conclusion of each season, specifically in February, May, and November. 
The surge in February may be attributed to heightened demand for printers and related services during the tax season, where businesses and individuals require printing services for financial documentation. May, characterized by graduation ceremonies, could contribute to an upswing in revenue. The increase in November is a result of holiday season preparations, heightened shopping activity, and promotional events leading to an increased demand for printed materials, culminating in a revenue peak.

# Pareto Chart of Movie Goers' Complaints

The goal for this visualisation part is to create a Pareto chart. A Pareto is a specific case of a bar and line chart. The specificity comes from the fact that the bars on the Pareto bar chart are displayed in decreasing order of frequency. The line chart on the other hand always shows the cumulative frequency on a Pareto. Cumulative frequency means the sum of frequencies occurring up to this point. So, for the first element, it will show the frequency of occurrence for the first element only. For the second element it will be the sum of frequency for the first and second element, and so on. The last cumulative frequency is 100% as it is the sum of all individual frequencies.

Related to the Pareto chart is the Pareto principle, also known as the 80/20 rule. Observed for the first time by Italian economist Wilfredo Pareto, who realized that around 80% of the land in Italy was owned by about 20% of the people.

This data set contains the top 5 Movie Goers complaints and the goal is to create a Pareto chart, based on the data. 

```python
df_movie_complaints = pd.read_csv(r"bar_line_chart.csv")
```

```python
df_movie_complaints
```

```python
fig, ax = plt.subplots(figsize = (10, 7))
ax.bar(df_movie_complaints["Movie Theater Goers Complaints"],
       df_movie_complaints["Number of complaints"],
       color = "dimgrey")
ax.set_ylabel("Number of Complaints",
              color = "dimgrey",
              weight = "bold")
ax.tick_params(axis = "y", 
               width = 2, 
               labelsize = "large",
               colors = "dimgrey")
ax1 = ax.twinx()
ax1.set_ylim(0, 1.1)
ax1.yaxis.set_major_formatter(PercentFormatter(xmax = 1.0))
ax1.plot(df_movie_complaints["Movie Theater Goers Complaints"].values,
         df_movie_complaints["frequency"].values,
         color = "r", 
         marker = "D")
ax1.set_ylabel("Frequency", 
               color = "r", 
               weight = "bold")
ax1.tick_params(axis = "y", 
                colors = "r", 
                width = 2, 
                labelsize = "large")
ax.set_title("Top 5 Movie Goers Complaints Frequency Distribution", fontsize = "14", weight = "bold")
plt.show()
```

In according to this frequency distribution we might notice that the top two complaints make up around 90% of the overall complaints. If the theaters address the first two or even the first complaint only for the poor projection quality, the number of complaints will be reduced drastically.
